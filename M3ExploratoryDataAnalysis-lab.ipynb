## Distribution


### Determine how the data is distributed


The column `ConvertedComp` contains Salary converted to annual USD salaries using the exchange rate on 2019-02-01.

This assumes 12 working months and 50 working weeks.


Plot the distribution curve for the column `ConvertedComp`.

```python
# Plot the distribution curve for the column `ConvertedComp`.
df = df.dropna(subset=['ConvertedComp'])

plt.figure(figsize=(10, 6))
plt.hist(df['ConvertedComp'], bins=30, edgecolor='k', alpha=0.7)
plt.title('Distribution of Converted Compensation')
plt.xlabel('Converted Compensation')
plt.ylabel('Frequency')
plt.grid(True)

plt.show()
```

Plot the histogram for the column `ConvertedComp`.


```python
plt.hist(df['ConvertedComp'], bins=20, edgecolor='k', alpha=0.7)
plt.xlabel('Salary (USD)')
plt.ylabel('Number of Respondents')
plt.title('Salary Distribution')

plt.show()
```

```python
man_only_count = df[df['Gender'] == "Man"].shape[0]

print(f"Number of responders who identified themselves only as 'Man': {man_only_count}")

woman_only_count = df[df['Gender'] == "Woman"].shape[0]

print(f"Number of responders who identified themselves only as 'Woman': {woman_only_count}")

woman_only_count_median = df[df['Gender'] == "Woman"]['ConvertedComp'].median()
print(f"Median: {woman_only_count_median}")
```
